## Introduction

This module presents an automated way to collect, analysis, and display data from the Scholarly Commons data source for the Children's Hospital of Philadelphia's Academic Entrepreneurship textbook.

Below are some methods the module provides

## Module Initialization

```python
from GUIModule import *

g = GUIModule()
```

## Total Downloads Over Time

```python
g.plotDownloadsOverTime(cumulative=True)
g.plotDownloadsOverTime()
```

## Number of Downloads by Country Over Time

The top 10 countries are by default displayed, with the remaining grouped under "other". More or less countries can be displayed using the cutoff parameter. Countries are sorted top to bottom in terms of download frequency. The bottom line is the most frequent country's line.

```python
g.plotDownloadsByCountryOverTime()
g.plotDownloadsByCountryOverTime(cutoff=5) #Cutoff set to 5 countries + others

g.plotDownloadsByCountryOverTime(cutoff=5,cumulative=True)
```
