## Introduction

This module presents an automated way to collect, analysis, and display data from the Scholarly Commons data source for the Children's Hospital of Philadelphia's Academic Entrepreneurship textbook.

Below are some methods the module provides

## Module Initialization

```python
from Dataset import *
import matplotlib.pyplot as plt

def cumulativeFreq(freq):
    a = []
    c = []
    for i in freq:
        a.append(i+sum(c))
        c.append(i)
    return np.array(a)

def column(a,colIndex):
    l = []
    for row in a:
        l.append(row[colIndex])
    return l

d = Dataset()
```

## Total Downloads Over Time

```python
times = [0,1,2,3,4]

plt.plot(times,cumulativeFreq(d.totalDownloadsOverTime()))
plt.ylabel('Cumulative Downloads')
plt.xlabel('Months Oct 2019 - Feb 2020')
plt.show()

plt.plot(times,d.totalDownloadsOverTime())
plt.ylabel('NonCumulative Downloads')
plt.xlabel('Months Oct 2019 - Feb 2020')
plt.show()
```

## Number of Downloads by Country Over Time

The top 10 countries are by default displayed, with the remaining grouped under "other". More or less countries can be displayed using the cutoff parameter. Countries are sorted top to bottom in terms of download frequency. The bottom line is the most frequent country's line.

```python
countries = d.totalDownloadsByCountryOverTime()

c = []
for i in range(len(countries[0])):
    if i == 0:
        c = cumulativeFreq(column(countries[1],i))
    else:
        c += cumulativeFreq(column(countries[1],i))
    plt.plot(times,c,label=countries[0][i])
plt.xlabel('Months Oct 2019 - Feb 2020')
plt.legend()
plt.show()

countries = d.totalDownloadsByCountryOverTime(5)

c = []
for i in range(len(countries[0])):
    if i == 0:
        c = cumulativeFreq(column(countries[1],i))
    else:
        c += cumulativeFreq(column(countries[1],i))
    plt.plot(times,c,label=countries[0][i])
plt.xlabel('Months Oct 2019 - Feb 2020')
plt.legend()
plt.show()
```

## Number of Downloads by Article in Aggregate
The top 10 articles are by default displayed, with the remaining grouped under "other". Below is listed the top 50 most downloaded articles.

```python
articles = d.totalDownloadsByArticle(50)
names = []
vals = []
for k,v in sorted(articles.items(),key=lambda item: item[1]):
    names.append(k)
    vals.append(v)
names.reverse()
y_pos = np.arange(len(names))
plt.bar(y_pos,vals,align='center',alpha=0.5)
plt.ylabel("Downloads")
plt.show()

for n in names:
    print(n)
```
